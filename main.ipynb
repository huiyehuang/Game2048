```python
from game_functions import initialize_grid, add_new_two, move_cells, check_game_over, print_grid

def main():
    grid = initialize_grid()
    while True:
        print_grid(grid)
        try:
            move = input("Enter move (w, a, s, d or 'q' to quit): ").lower()
            if move not in ['w', 'a', 's', 'd', 'q']:
                raise ValueError("Invalid move. Please enter 'w', 'a', 's', 'd', or 'q' to quit.")
            if move == 'q':
                print("Game exited by user.")
                break
            new_grid = move_cells(grid, move)
            if new_grid != grid:
                grid = new_grid
                add_new_two(grid)
            if check_game_over(grid):
                print("Game Over!")
                break
        except ValueError as e:
            print(e)

if __name__ == "__main__":
    main()
```
